```python
import numpy as np
import qutip as q
import matplotlib.pyplot as plt
import astropy.units as u
import astropy.constants as c
from math import factorial
import scipy.linalg as lin
```

```python
sub_t = 350*u.um #substrate thickness (Si in this case)
metal_t = 250*u.nm #Deposited metal thickness (Al)
Sc = 67*u.fF/(u.um)**2 #JJ specific capacitance
epsilon = 11.45
W_jj = 200*u.nm #junction width
phi0 = c.h/2/c.e.si
T = 30*u.mK
```

```python
def get_qubit_Es(Cq,Lj):
    Ec = (c.e.si**2/2/Cq).to(u.J)
    Ej = ((phi0/2/np.pi)**2/Lj).to(u.J)
    return (Ec, Ej)
    
```

```python
C1 = 115*u.fF
C2 = 115*u.fF
# C2 = 80*u.fF
# Cg = 0*u.fF
Cg = 6*u.fF

C_mat = np.array([[C1.value+Cg.value,-Cg.value],[-Cg.value, C2.value+Cg.value]])
C_mat_inv = lin.inv(C_mat)
C_mat_inv /= u.fF
Ec1 = (c.e.si**2/2*C_mat_inv[0,0]).to(u.J)
Ec2 = (c.e.si**2/2*C_mat_inv[1,1]).to(u.J)
Ecg = (c.e.si**2/2*C_mat_inv[1,0]).to(u.J)


Lj = 13*u.nH
Ej = ((phi0/2/np.pi)**2/Lj).to(u.J)
Ej1 = Ej
Ej2 = Ej

# Ecg = 0*u.J
# Ecg = (c.e.si**2/2/Cg).to(u.J)

nz1 = 1j/2*(Ej1/2/Ec1)**(1/4)
nz2 = 1j/2*(Ej2/2/Ec2)**(1/4)

pz1 = (2*Ec1/Ej1)**(1/4)
pz2 = (2*Ec2/Ej2)**(1/4)
```

```python
N = 5
I1 = q.Qobj([[1,0],[0,0]])
I2 = q.Qobj([[0,0],[0,1]])
I = q.qeye(N)
Id1 = np.array(q.tensor(I2, I))
Id2 = np.array(q.tensor(I1, I))
```

```python
qc1 = q.tensor(I1, q.create(N))
qd1 = q.tensor(I1, q.destroy(N))
qc2 = q.tensor(I2, q.create(N))
qd2 = q.tensor(I2, q.destroy(N))
```

```python
qd1*qc1+qd2*qc2
```

```python
n1 = nz1*np.array(qc1-qd1)
n1d = np.transpose(np.conj(n1))
n2 = nz2*np.array(qc2-qd2)
n2d = np.transpose(np.conj(n2))

phi1 = pz1*np.array(qc1+qd1)
phi1d = np.transpose(np.conj(phi1))
phi2 = pz2*np.array(qc2+qd2)
phi2d = np.transpose(np.conj(phi2))
```

```python
H1 = 4*Ec1*n1d@n1+1/2*Ej1*phi1d@phi1-1/factorial(4)*Ej1*(phi1d@phi1)@(phi1d@phi1)
H2 = 4*Ec2*n2d@n2+1/2*Ej2*phi2d@phi2-1/factorial(4)*Ej2*(phi2d@phi2)@(phi2d@phi2)
H_int = 4*Ecg/2*((n1d+Id1)@(n2+Id2)+(n2d+Id2)@(n1+Id1))
```

```python
H_tot = H1+H2+H_int
```

```python
h = np.array(H_tot)
```

```python
eig, vec = lin.eigh(h)
```
